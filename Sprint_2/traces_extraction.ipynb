```python
import pm4py
import os 
import pandas as pd

# CHANGE THE PATH TO YOU DATA FODER WITH ORIGINAL DATA - XES.GZ
path_to_data_foder = '/Users/alexraudvee/Desktop/TU_e/year_2/Q3Y2_Process_Mining/process_mining_data'

# HERE YOU HAVE TO CHOOSE THE DATASET WITH WHICH YOU WANNA WORK "BPI_Challenge_2012" OR "BPI_Challenge_2017"
chosed_dataset = 'BPI_Challenge_2012'
```

```python
if __name__ == '__main__':
    dataframe_train = pd.read_csv(f'../data/train_{chosed_dataset}.csv')
    dataframe_test = pd.read_csv(f'../data/test_{chosed_dataset}.csv')
    dataframe_train_test = pd.read_csv(f'../data/train_test_{chosed_dataset}.csv')

    dataframe_train = pm4py.format_dataframe(dataframe_train, case_id='case:concept:name', activity_key='concept:name', timestamp_key='time:timestamp')
    dataframe_test = pm4py.format_dataframe(dataframe_test, case_id='case:concept:name', activity_key='concept:name', timestamp_key='time:timestamp')
    dataframe_train_test = pm4py.format_dataframe(dataframe_train_test, case_id='case:concept:name', activity_key='concept:name', timestamp_key='time:timestamp')

    event_log_train = pm4py.convert_to_event_log(dataframe_train)
    event_log_test = pm4py.convert_to_event_log(dataframe_test)
    event_log_train_test = pm4py.convert_to_event_log(dataframe_train_test)
    
    pm4py.write_xes(event_log_train, f'../data/train_{chosed_dataset}.xes')
    pm4py.write_xes(event_log_test, f'../data/test_{chosed_dataset}.xes')
    pm4py.write_xes(event_log_train_test, f'../data/train_test_{chosed_dataset}.xes')

    event_logs_files_names = [f'train_{chosed_dataset}', f'test_{chosed_dataset}', f'train_test_{chosed_dataset}']
```

```python
for event_log_file in event_logs_files_names:

    log = pm4py.read_xes(os.path.join(f"../data/{event_log_file}.xes"))
    log = pm4py.convert_to_event_log(log)

    trace_df_list = []
    for trace in log:
        case_concept_name = trace.attributes['concept:name']
        concept_name_col = []
        timestamp_col = []
        
        for event in trace:
            concept_name_col.append(event['concept:name'])
            timestamp_col.append(event['time:timestamp'])
        
        df_trace = pd.DataFrame({'concept:name': concept_name_col, 'time:timestamp': timestamp_col})
        df_trace.name = f'{case_concept_name}'

        trace_df_list.append(df_trace)

    df_final = pd.DataFrame({'case:concept:name': [int(df.name) for df in trace_df_list], 'trace': trace_df_list})

    df_final.to_csv(f'../data/traces_{event_log_file}.csv')

    if os.path.exists(f'../data/{event_log_file}.xes'):
        # Remove the file
        os.remove(f'../data/{event_log_file}.xes')
    
df_final
```
